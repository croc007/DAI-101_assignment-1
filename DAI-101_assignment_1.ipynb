```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/Banking_churn_prediction.csv"
df = pd.read_csv(file_path)


# 1. Variable Identification & Typecasting

print("\n### Variable Identification & Typecasting ###\n")

# Display column names and data types
print("Data Types:\n", df.dtypes)

# Convert categorical variables to category type
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category')


# 2. Univariate Analysis: Numerical Variables

print("\n### Univariate Analysis: Numerical Variables ###\n")
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    print(f"\nSummary Statistics for {col}:")
    print(df[col].describe())

    # Plot histogram and boxplot
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(df[col], kde=True, ax=axes[0], bins=30)
    axes[0].set_title(f"Histogram of {col}")

    sns.boxplot(x=df[col], ax=axes[1])
    axes[1].set_title(f"Boxplot of {col}")

    plt.show()


# 3. Univariate Analysis: Categorical Variables

print("\n### Univariate Analysis: Categorical Variables ###\n")
cat_cols = df.select_dtypes(include=['category']).columns

for col in cat_cols:
    print(f"\nValue Counts for {col}:")
    print(df[col].value_counts())

    # Plot bar chart
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[col], data=df, palette="viridis")
    plt.title(f"Bar Chart of {col}")
    plt.xticks(rotation=45)
    plt.show()


# 4. Univariate Analysis: Missing Values

print("\n### Univariate Analysis: Missing Values ###\n")
missing_values = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({"Missing Values": missing_values, "Percentage": missing_percent})
print(missing_df[missing_df["Missing Values"] > 0])


# 5. Univariate Analysis: Outlier Detection

print("\n### Univariate Analysis: Outlier Values ###\n")

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"\n{col}: {len(outliers)} outliers detected")


# 6. Summary of Univariate Analysis

print("\n### Summary of Univariate Analysis ###\n")
print(f"Total Rows: {df.shape[0]}, Total Columns: {df.shape[1]}")
print(f"Numerical Variables: {len(num_cols)}, Categorical Variables: {len(cat_cols)}")
print(f"Missing Values Found: {missing_df[missing_df['Missing Values'] > 0].shape[0]}")
print(f"Outliers Detected in {sum(len(df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]) > 0 for col in num_cols)} columns")
```
